# filling df2 missing values

```python
for i in ['attacktype2', 'attacktype3']:
    df2[i].fillna(9, inplace = True)
```

```python
for i in ['targsubtype1', 'targsubtype2', 'targsubtype3']:
    df2[i].fillna(114, inplace = True)
```

```python
for i in ['targtype2', 'targtype3', 'targtype1']:
    df2[i].fillna(24, inplace = True)
```

```python
for i in ['claimmode', 'claimmode2', 'claimmode3']:
    df2[i].fillna(12, inplace = True)
```

```python
for i in ['weapsubtype1', 'weapsubtype2', 'weapsubtype3', 'weapsubtype4']:
    df2[i].fillna(40, inplace = True)
```

```python
for i in ['weaptype2', 'weaptype3', 'weaptype4']:
    df2[i].fillna(15, inplace = True)
```

```python
for i in ['propextent']:
    df2[i].fillna(5, inplace = True)
```

```python
for i in ['hostkidoutcome']:
    df2[i].fillna(10, inplace = True)
```

```python
#columns filled with their mean values
mean_col = ['latitude', 'longitude', 'specificity', 'doubtterr', 'multiple',
            'guncertain1', 'guncertain2', 'guncertain3', 'number of perpetrators',
            'nperpcap', 'kill', 'nkillus', 'nkillter', 'wounded', 'nwoundus',
            'nwoundte', 'ishostkid', 'nhostkid', 'nhostkidus', 'hours',
            'days', 'ransom', 'ransomamt', 'ransomamtus', 'ransompaid',
            'ransompaidus', 'released', 'claimed', 'claim2','claim3',
           'compclaim', 'propvalue']
for i in mean_col:
    df2[i].fillna(df2[i].mean(axis = 0), inplace = True)
```
