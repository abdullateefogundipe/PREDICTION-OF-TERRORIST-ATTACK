```python
# Function for the creation of the date column, and correction of wrong day or month
def make_date(df):
    new_date = ""
    year = df['Year']
    month = df['Month']
    day = df['Day']
    
    if month < 1:
        month = 1
        
    if day < 1:
        day = 1
    
    new_date = f"{year}-{month}-{day}"
    
    return new_date
```

```python
df = df.assign(
    Date=df.apply(make_date, axis=1)
)
df.Date = pd.to_datetime(df.Date)
```
